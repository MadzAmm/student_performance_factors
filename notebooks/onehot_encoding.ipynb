```python
# 1. import class onehotencoder dari library sklearn

from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv ('../data/spf.csv')

df.head(10)
```

```python
# Rubah nilai dari varabel yang kita ingin onehot encoding dari data kategotik menjadi data numerik dengan class yang telah dipanggil di atas


y = OneHotEncoder(sparse_output=False)
y = y.fit_transform(df[['Teacher_Quality']])
y

```

```python
# rubah array diatas menjadi data fram dengan menggunakana library pandas

df_y =  pd.DataFrame(y) #--->baris kode yang merubah jadi data frame
df_y.head(10) #---> baris kode yang menampilkan data frame
```

```python
# header = df['Teacher_Quality'].sort_values()
# df_y.columns = 'Quality '+header
# df_y

# coba kita cocokan posisi hasil onehotencoding dengan posisi value katedorik yang dimiliki oleh atribut teacher quality

# 1. TENTUKAN NAMA KATEGORI YANG BENAR
# INI SANGAT PENTING: Ganti urutan ['High', 'Low', 'Medium']
# agar sesuai dengan urutan asli kolom 0, 1, 2 di df_y
# (Jika salah urut, datanya akan salah pemetaan)
nama_kategori_yang_benar = ['High', 'Low', 'Medium'] 
 
# 2. Terapkan nama header baru ini ke df_y
df_y.columns = ['Quality_' + kat for kat in nama_kategori_yang_benar]
 
# 3. Tampilkan df_y yang sudah memiliki header yang benar
print(df_y.head(10))
```

```python
# kita cocokan dengan data kategorik teacher quality yang ada pada data frame awal apakah hasil ohe ini sesuai dengan data ketegorik sebelumnya (pastinya sesuai) dengan menggunakan concat. kenapa concat? karena kita hanya mengambil salah satu atribut/variable/kolo atau satu series yang ada pada df lalu digabungkan dengan tabel yang baru df_y. kalau mau menggabungkan satu data frame maka gunakan .join.

# 1. Ambil kolom asli DAN BUAT ULANG INDEKSNYA
original_col = df['Teacher_Quality'].reset_index(drop=True) #--->original kolom teacher quality pada df awal

# 2. Ambil data one-hot DAN BUAT ULANG INDEKSNYA
encoded_cols = df_y.reset_index(drop=True) #---> ini data hasil ohe yang hendak kita gabungkan 

# kenapa perlu reset index terlebih dahulu sebelum concat?, karena df.join() dan pd.concat() bekerja dengan cara yang sama. Keduanya menggunakan indeks untuk mencocokkan baris, maka HARUS menggunakan reset_index(drop=True) pada kedua bagian sebelum menggabungkannya. Ini adalah satu-satunya cara agar datanya cocok dan tidak NaN.

# 3. GABUNGKAN (Gunakan pd.concat, BUKAN .join)
# pd.concat dirancang untuk menggabungkan Series 
# axis=1 berarti menggabungkan secara horizontal (berdampingan)
df_concat = pd.concat([original_col, encoded_cols], axis=1)

# 4. Tampilkan hasilnya (Sekarang akan berhasil dan tanpa NaN)
print(df_concat.head(10))
```
